```python
punctuation_marks = ['.', ',', ';', ':', '?', '!', '...', '\'', '/', '(', ')', '-','"']
```

```python
datas = []
```

```python
with open("/content/drive/MyDrive/SIC2025 - 404 Not Found/Crawl Dataset/input.txt", "r") as f:
  for sentence in f:
    customized_sentence = sentence.lower()
    for mark in punctuation_marks:
      customized_sentence = customized_sentence.replace(mark, '')
      sentence = " ".join(sentence.replace(mark, ' ' + mark + ' ').split())
    data = {
        "original_text": customized_sentence.strip(),
        "corrected_text": sentence.strip()
    }
    datas.append(data)
```

```python
import json
with open("/content/drive/MyDrive/SIC2025 - 404 Not Found/Crawl Dataset/dataset.json", "w") as f:
  json.dump(datas, f, ensure_ascii=False, indent=2)
```
